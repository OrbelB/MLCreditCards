```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("./clean_data/creditcard_application_clean_data.csv") 
# reading cleaned data
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df
```

```python
sns.histplot(data=df, x="AGE_YEARS", bins=20, stat='proportion').set(title='Age proportions')
```

```python
sns.histplot(data=df, x="EMPLOYMENT_LENGTH", bins=20, stat='proportion').set(title='Employment length proportions')
```

```python
sns.histplot(data=df, x="ANNUAL_INCOME", bins=30, stat='proportion').set(title='Anuual income proportions')
```

```python
vplot = sns.violinplot(data=df, x="ANNUAL_INCOME")
vplot.set(ylabel='Counts',title='Annual income counts') # setting title and y-label
```

```python
sns.boxplot(data=df, x="ANNUAL_INCOME", y='NAME_HOUSING_TYPE', orient='h').set(title='Anuual income vs housing type')
#if we dont mention the orient='h' attribute, the graph results are plotted vertically, instead of horizontal.
```

```python
sns.violinplot(data=df, x="ANNUAL_INCOME", y='NAME_EDUCATION_TYPE', orient='h').set(title='Annual income vs Education type')
```

```python
sns.scatterplot(data=df, x="AGE_YEARS", y='ANNUAL_INCOME').set(title='Age vs Anuual income')
```

```python
df['NAME_EDUCATION_TYPE'].value_counts()
```

```python
ax = plt.figure(figsize=(10,12))
ax.set_facecolor((1, 1, 1))

a = (df['NAME_EDUCATION_TYPE']==4).sum()
b = (df['NAME_EDUCATION_TYPE']==1).sum()
c = (df['NAME_EDUCATION_TYPE']==2).sum()
d = (df['NAME_EDUCATION_TYPE']==3).sum()
e = (df['NAME_EDUCATION_TYPE']==0).sum()

piedata = np.array([a, b, c, d, e])
colors = sns.color_palette('hls',8)
labels = ['4', '1', '2', '3' , '0']
plt.pie(piedata, labels = labels, colors=colors)
plt.legend(title = 'Education types')
plt.title("Education")
```

```python
ax = plt.figure(figsize=(10,12))
ax.set_facecolor((1, 1, 1))

a = (df['NAME_HOUSING_TYPE']==1).sum()
b = (df['NAME_HOUSING_TYPE']==5).sum()
c = (df['NAME_HOUSING_TYPE']==2).sum()
d = (df['NAME_HOUSING_TYPE']==4).sum()
e = (df['NAME_HOUSING_TYPE']==3).sum()
f = (df['NAME_HOUSING_TYPE']==0).sum()

piedata = np.array([a, b, c, d, e, f])
colors = sns.color_palette('hls',8)
labels = ['1', '5' ,'2' ,'4' ,'3', '0']
plt.pie(piedata, labels = labels, colors=colors)
plt.legend(title = 'Housing types')
plt.title("Housing")
```
