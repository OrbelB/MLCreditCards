```python
ax = plt.figure(figsize=(10,12))
ax.set_facecolor((1, 1, 1))

a = data['NAME_EDUCATION_TYPE']==0
a = a.sum()
b = data['NAME_EDUCATION_TYPE']==1
b = b.sum()
c = data['NAME_EDUCATION_TYPE']==2
c = c.sum()
d = data['NAME_EDUCATION_TYPE']==3
d = d.sum()
e = data['NAME_EDUCATION_TYPE']==4
e = e.sum()

piedata = np.array([a, b, c, d, e])
colors = sns.color_palette('hls',8)
labels = ['0', '1', '2', '3' , '4']
plt.pie(piedata, labels = labels, colors=colors)
plt.legend(title = 'Education types')
```

```python
ax = plt.figure(figsize=(10,12))
ax.set_facecolor((1, 1, 1))

a = data['NAME_HOUSING_TYPE']==0
a = a.sum()
b = data['NAME_HOUSING_TYPE']==1
b = b.sum()
c = data['NAME_HOUSING_TYPE']==2
c = c.sum()
d = data['NAME_HOUSING_TYPE']==3
d = d.sum()
e = data['NAME_HOUSING_TYPE']==4
e = e.sum()
f = data['NAME_HOUSING_TYPE']==5
f = f.sum()
piedata = np.array([a, b, c, d, e, f])
colors = sns.color_palette('hls',8)
labels = ['0', '1', '2' ,'3' ,'4' ,'5']
plt.pie(piedata, labels = labels, colors=colors)
plt.legend(title = 'Housing types')
```
